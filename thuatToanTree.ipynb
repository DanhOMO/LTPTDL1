```python
# 1. Import các thư viện cần thiết
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn import tree
import matplotlib.pyplot as plt

# 2. Tải bộ dữ liệu Iris
iris = load_iris()
X = iris.data  # Đặc trưng (features)
y = iris.target  # Nhãn (labels)

# 3. Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 4. Xử lý giá trị thiếu (nếu có)
# Mặc dù bộ dữ liệu Iris không có giá trị thiếu, nhưng nếu sử dụng bộ dữ liệu khác, bạn có thể làm như sau:
# imputer = SimpleImputer(strategy='mean')
# X_train = imputer.fit_transform(X_train)
# X_test = imputer.transform(X_test)

# 5. Chuẩn hóa dữ liệu (nếu cần thiết)
# Dù Decision Tree không yêu cầu chuẩn hóa dữ liệu, nhưng nếu bạn muốn chuẩn hóa cho các thuật toán khác, bạn có thể làm như sau:
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 6. Khởi tạo và huấn luyện mô hình Decision Tree
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train_scaled, y_train)

# 7. Dự đoán và đánh giá mô hình
y_pred = clf.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)

print("Độ chính xác của mô hình Decision Tree:", accuracy)
print("Báo cáo phân loại:\n", classification_report(y_test, y_pred))

# 8. Vẽ cây quyết định (Decision Tree)
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.title("Decision Tree for Iris Classification")
plt.show()
```
