```python
# transformation techinque gồmm binning , (onehot endcoding and functions ) data preparetion 
# binning : chia dữ liệu thành các khoảng dữ liệu
# onehot encoding : chuyển dữ liệu thành dạng số
# functions : chuyển dữ liệu thành dạng số
# data preparetion : chuẩn hóa dữ liệu
```

```python

df['bin1'] = pd.qcut(df['Math'], q=5)
df
# chia dữ liệu thành 5 khỏang 
```

```python
df['bin1'].value_counts()
# đếm số lựogn pt tỏng môix khoảnng
```

```python
df['bin2'] = pd.qcut(df['Math'], q=[0, .2, .4, .6, .8, 1])
df
df['bin2'].value_counts()
# đếm số lựogn pt tỏng môix khoảnng

# chia dữ liệu thành 5 khoảng với tỉ lệ khác nhau
```

```python
df['bin3'] = pd.qcut(df['Math'],
                           q=[0, .2, .4, .6, .8, 1],
                           labels=['A', 'B', 'C', 'D', 'E'])
df
# chia dữ liệu thành 5 khoảng với tỉ lệ khác nhau và đặt tên cho mỗi khoảng 
```

```python
df['bin3'].value_counts()
# đếm số lựogn pt tỏng môix khoảnng
```

```python

df['bin4'] = pd.cut(df['Eng'],
                      bins=[0, 5, 8, float('Inf')],
                      labels=['Bad', 'OK', 'Good'])
df
# chia dữ liệu thành 3 khoảng và đặt tên cho mỗi khoảng 
```

```python
df['bin4'].value_counts()
# đếm số lựogn pt tỏng môix khoảnng
```

```python
# function : non malization, standandization, popuba function
import pandas as pd

year = [2010,2011,2012,2013,2014,2015]

salaryHCM = [8.5,9.0,10.5,12.0,11.0,10.0]

salaryLA = [5.5,6.0,6.0,5.5,6.0,5.0]

df = pd.DataFrame(list(zip(year, salaryHCM,salaryLA)),
                  columns=['Year', 'HCM','LA'])
df
```

```python
# max absolute
max_la = df['LA'].abs().max()
df['macLA'] = df['LA'] / max_la
df
```

```python

max_la = df['HCM'].abs().max()
df['macHCM'] = df['HCM'] / max_la

df
```

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)

```

```python
vSc = scaler.transform([[4.5,5.75]])
vSc
# chuẩn hóa dữ liệu 4.5,5.75 theo dữ liệu đã có 
```

```python
vR = scaler.inverse_transform([[0.5,0.8]])
vR
```

```python
# min max scaler 
min_la = df['LA'].min()
max_la = df['LA'].max()
df['mmcLA'] = (df['LA'] - min_la) / (max_la - min_la)

df
```
